```python
import pandas as pd
```

```python
layout_dir = 'LH-Upimium/2 Warehouse Layout/'
data_dir = 'LH-Upimium/ob_stats_20200204b/'
```

```python
df_multi_batch_pick_task = pd.read_csv(data_dir + 'multi_batch_pick_task.csv', names=['wave', 'sub-wave', 'created', 'zone', 'bin', 'item', 'order_qty', 'pick_qty', 'status'])
df_multi_batch_pick_task.created = pd.to_datetime(df_multi_batch_pick_task.created)
df_multi_batch_pick_task = df_multi_batch_pick_task.sort_values(['created']).reset_index(drop=True)
df_multi_batch_pick_task
```

```python
df_multi_batch_pick_transaction_log = pd.read_csv(data_dir + 'multi_batch_pick_transaction_log.csv', names=['picked', 'user', 'subwave', 'wave', 'wave_created', 'zone', 'bin', 'item', 'qty'])
df_multi_batch_pick_transaction_log.wave_created = pd.to_datetime(df_multi_batch_pick_transaction_log.wave_created)
df_multi_batch_pick_transaction_log.picked = pd.to_datetime(df_multi_batch_pick_transaction_log.picked)
df_multi_batch_pick_transaction_log = df_multi_batch_pick_transaction_log.sort_values(['wave_created']).reset_index(drop=True)
df_multi_batch_pick_transaction_log
```

```python
df_multi_carton_pick_task = pd.read_csv(data_dir + 'multi_carton_pick_task.csv', names=['wave', 'created', 'zone', 'bin', 'item', 'order_qty', 'pick_qty', 'status'])
df_multi_carton_pick_task.created = pd.to_datetime(df_multi_carton_pick_task.created)
df_multi_carton_pick_task = df_multi_carton_pick_task.sort_values(['created']).reset_index(drop=True)
df_multi_carton_pick_task
```

```python
df_multi_carton_pick_transaction_log = pd.read_csv(data_dir + 'multi_carton_pick_transaction_log.csv', names=['picked', 'user', 'order', 'zone', 'bin', 'item', 'qty'])
df_multi_carton_pick_transaction_log.picked = pd.to_datetime(df_multi_carton_pick_transaction_log.picked)
df_multi_carton_pick_transaction_log = df_multi_carton_pick_transaction_log.sort_values(['picked']).reset_index(drop=True)
df_multi_carton_pick_transaction_log
```

```python
df_order_line_detail = pd.read_csv(data_dir + 'order_line_detail.csv', names=['order', 'freight', 'order_type', 'status', 'subwave', 'created', 'wave', 'item', 'qty_ordered', 'qty_picked', 'qty_packed', 'qty_shipped'])
df_order_line_detail.created = pd.to_datetime(df_order_line_detail.created)
df_order_line_detail = df_order_line_detail.sort_values(['created']).reset_index(drop=True)
df_order_line_detail
```

```python
df_order_milestone = pd.read_csv(data_dir + 'order_milestone.csv', names=['order', 'freight', 'order_type', 'status', 'subwave', 'created', 'wave', 'imported', 'released', 'picked', 'packed', 'shipped', 'cancel'])
df_order_milestone.created = pd.to_datetime(df_order_milestone.created)
df_order_milestone.imported = pd.to_datetime(df_order_milestone.imported)
df_order_milestone.released = pd.to_datetime(df_order_milestone.released)
df_order_milestone.picked = pd.to_datetime(df_order_milestone.picked)
df_order_milestone.packed = pd.to_datetime(df_order_milestone.packed)
df_order_milestone.shipped = pd.to_datetime(df_order_milestone.shipped)
df_order_milestone.cancel = pd.to_datetime(df_order_milestone.cancel)
df_order_milestone = df_order_milestone.sort_values(['created']).reset_index(drop=True)
df_order_milestone
```
